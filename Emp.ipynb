```python
class Employee:
    def __init__(self):
        print("Fill your information:")
        self.user_details = {}
        self.user_skill = []
        self.user_information()

    def user_information(self):
        name = input("Enter your name: ")
        self.user_details["name"] = name

        # Email validation
        while True:
            email = input("Enter your correct email: ")
            if email.endswith("@gmail.com"):
                self.user_details["email"] = email
                break
            else:
                print("Invalid email")

        # Phone number validation
        while True:
            phone_number = input("Enter your 10-digit phone number: ")
            if phone_number.isdigit() and len(phone_number) == 10:
                self.user_details["Phone number"] = phone_number
                break
            else:
                print("Invalid phone number")

        # Skill input
        while True:
            skill = input("Enter your skill (or type 'No' to stop): ")
            if skill.lower() == "no":
                break
            else:
                self.user_skill.append(skill.lower())

        self.user_details["skill"] = self.user_skill
        self.employee_skill()

    def employee_skill(self):
        job_roles = input("""
Choose your job role:
1. Apply for Python Developer
2. Apply for Data Analytics
3. Exit
""")

        if job_roles == "1":
            self.user_details["apply_for"] = "Python Developer"
            self.check_emp_skill()
        elif job_roles == "2":
            self.user_details["apply_for"] = "Data Analytics"
            self.check_emp_skill()
        elif job_roles == "3":
            print("Bye bye.")
        else:
            print("Invalid choice")
            self.employee_skill()  # Retry
    def check_emp_skill(self):
        python_developer_skills = {"python", "html", "css", "js", "sql"}
        data_analytics_skills = {"python", "oops", "numpy", "pandas", "nlp", "excel"}
        user_skills = set(self.user_details["skill"])
    
        applied_role = self.user_details["apply_for"]
    
        if applied_role == "Python Developer":
            required = python_developer_skills
        elif applied_role == "Data Analytics":
            required = data_analytics_skills
        else:
            print("Invalid role.")
            return
    
        print(f"Your Skills: {user_skills}")
        print(f"Required Skills: {required}")
    
        # Check if all required skills are present
        if required.issubset(user_skills):
            print(f"You are Selected for {applied_role}")
        else:
            print(f"Not Selected for {applied_role}. You are missing some required skills.")

e = Employee()
```
